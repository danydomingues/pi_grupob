# Download e Armazenamento - Arquivos de SP

```
import urllib

url_estadual = {
  #CSV - Arquivos 'acidentes_nao_fatais.csv' e 'obitos.csv' via url
  "acidentes_nao_fatais":["http://painelderesultados.infosiga.sp.gov.br/bases/acidentes_naofatais.csv", "acidentes_nao_fatais.csv"],
  "obitos":["http://painelderesultados.infosiga.sp.gov.br/bases/obitos_publico.csv", "obitos.csv"],
  
  #XLSX
    "acidentes_fatais":["http://painelderesultados.infosiga.sp.gov.br/bases/acidentes_fatais.xlsx", "acidentes_fatais.xlsx"]
}

#CSv -  Envio dos arquivos para o Blob Storage arquivos 
urllib.request.urlretrieve(url_estadual["acidentes_nao_fatais"][0], "/dbfs/mnt/landing/web_scraping/estadual/raw/csv/{}".format(url_estadual["acidentes_nao_fatais"][1]))
urllib.request.urlretrieve(url_estadual["obitos"][0], "/dbfs/mnt/landing/web_scraping/estadual/raw/csv/{}".format(url_estadual["obitos"][1]))

#XLSX
urllib.request.urlretrieve(url_estadual["acidentes_fatais"][0], "/dbfs/mnt/landing/web_scraping/estadual/raw/xlsx/{}".format(url_estadual["acidentes_fatais"][1]))
```
